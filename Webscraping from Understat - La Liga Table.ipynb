```python
#import libraries
import requests
from bs4 import BeautifulSoup
import json
import pandas as pd
```

```python
#get the base url
url = 'https://understat.com/league/La_liga'
url
```

```python
webdata = requests.get(url)
```

```python
soup = BeautifulSoup(webdata.content,'lxml')
scripts = soup.find_all('script') ##this will find all the scripts in the URL
```

```python
scripts
```

```python
#get only the table

string = scripts[2].string ##turn data into a string
string
```

```python
#remove all the uneccesary symbols

index_start = string.index("('")+2 
index_end = string.index("')")
json_data = string[index_start:index_end] 
json_data = json_data.encode('utf8').decode('unicode_escape')
data = json.loads(json_data)
```

```python
data
```

```python
##turn data into a dataframe

df = pd.DataFrame(data.values())
```

```python
df
```

```python
#convert history column to rows
df = df.explode("history")
```

```python
df
```

```python
df = pd.DataFrame(data.values())
df = df.explode("history")
h = df.pop("history")
df = pd.concat([df.reset_index(drop=True), pd.DataFrame(h.tolist())], axis=1)
df = df.infer_objects()
```

```python
df
```

```python
# Get the unique values in the "home and away" column
unique_values = df['h_a'].unique()

# Create new columns for each unique value in the "h_a" column
for value in unique_values:
    df[value] = 0

# Loop through each row of the data
for index, row in df.iterrows():
    # Get the value of the "h_a" column for the current row
    h_a_value = row["h_a"]
    
    # Set the value of the new column for the current row
    df.at[index, h_a_value] = 1
```

```python
df
```

```python
#drop unnecessary columns
df.drop(columns=["ppda_allowed", "ppda"], axis=1, inplace=True)
```

```python
#Export raw data to a csv file
df.to_csv('LaLigaTeamStats.csv')
```
